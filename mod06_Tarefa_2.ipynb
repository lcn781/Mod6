```python
import pandas as pd
import requests
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


sinasc = pd.read_csv('SINASC_RO_2019.csv')
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python


plt.figure(figsize=(10, 6))

plt.boxplot(sinasc['IDADEMAE'].dropna(), vert=True, patch_artist=True, boxprops=dict(facecolor='lightblue'), medianprops=dict(color='red'))
plt.title('Boxplot da Idade da Mãe')
plt.ylabel('Idade')
plt.grid(True, alpha=0.3)

idade = sinasc['IDADEMAE'].dropna()
plt.text(0.7, 0.95, f'Média: {idade.mean():.1f}\nMediana: {idade.median():.1f}', transform=plt.gca().transAxes, bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))


plt.tight_layout()
plt.show()


```

####  2) Faça um boxplot do peso do bebê

```python
plt.figure(figsize=(10, 6))

plt.boxplot(sinasc['PESO'].dropna(), vert=True, patch_artist=True, boxprops=dict(facecolor='red'), medianprops=dict(color='blue'))
plt.title('Boxplot Peso ao Nascer')
plt.ylabel('Peso (gramas)')
plt.grid(True, alpha=0.3)

peso = sinasc['PESO'].dropna()
plt.text(0.7, 0.95, f'Média: {peso.mean()/1000:.2f} kg\nMediana: {peso.median()/1000:.2f} kg', transform=plt.gca().transAxes, bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))


plt.tight_layout()
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
plt.figure(figsize=(10, 6))

sinasc['IDADEMAE'].hist(bins=20, color='gray', edgecolor='black')
plt.title('Histograma Idade da Mãe')
plt.xlabel('Idade')
plt.ylabel('Frequência')

idade = sinasc['IDADEMAE'].dropna()
plt.text(0.7, 0.95, f'Média: {idade.mean():.1f}\nMediana: {idade.median():.1f}', transform=plt.gca().transAxes,bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

plt.tight_layout()
plt.show()

```

#### 4) Faça um histograma do peso do bebê

```python
# Versão minimalista
plt.figure(figsize=(10, 6))

# Histograma simples
sinasc['PESO'].hist(bins=30, color='gray', alpha=0.7, edgecolor='black')

plt.title('Histograma - Peso ao Nascer (PESO)')
plt.xlabel('Peso (gramas)')
plt.ylabel('Frequência')
plt.grid(True, alpha=0.3)

# Adicionar estatísticas básicas
peso = sinasc['PESO'].dropna()
plt.text(0.7, 0.95,
         f'Média: {peso.mean()/1000:.2f} kg\nMediana: {peso.median()/1000:.2f} kg',
         transform=plt.gca().transAxes,
         bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

plt.tight_layout()
plt.show()
```
